```python
import cv2
import numpy as np
```

# Reading image

```python
img=cv2.imread('download.jpg')
```

# Structuring element or kernal

```python
kernal=np.ones((3,3),np.uint8)
```

# Erosion

```python
erosion=cv2.erode(img,kernal,iterations=7)
```

# Display Erosion

```python
cv2.imshow('original',img)
cv2.imshow('erosion',erosion)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# dilation

```python
dilation=cv2.dilate(img,kernal,iterations=3)
```

```python
cv2.imshow('original',img)
cv2.imshow('erosion',erosion)
cv2.imshow('dilation',dilation)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Edge Detection for Video

```python
vid=cv2.VideoCapture("Magic Tree_2.mp4") 
while(1):
    ret,frame=vid.read()
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    cv2.imshow("Magic Tree_2.mp4",frame)
    edges=cv2.Canny(frame,100,200)
    cv2.imshow("edges",edges)
    if cv2.waitKey(5)& 0xFF==ord('a'):
        break
vid.release()
cv2.destroyAllWindows()
```

```python
#Image processing libary
from skimage.feature import hog
from skimage import data, exposure
```

# READ IMAGE

```python
image=cv2.imread('Cat03.jpg')
```

# HOG

```python
fd, hod_image = hog(image,orientations=9,pixels_per_cell=(8,8),cells_per_block=(2,2), visualize=True, channel_axis=2)
```

# RESCALING

```python
rescale=exposure.rescale_intensity(hod_image, in_range=(0,10))
```

# Plot

```python
figure,(a1,a2) =plt.subplots(1,2, figsize=(12,6), sharex=True, sharey=True)
```

```python
a1.axis('off')
a1.imshow(image)
a1.set_title('Input image')

a2.axis('off')
a2.imshow(rescale)
a2.set_title('Histrogram of Oriented Gradients')
plt.show()
```
