```python
gray=cv2.imread("explore.jpg",0)
cv2.imshow("gray",gray)
cv2.waitKey(0)
```

```python
img=cv2.imread("explore.jpg")
```

```python
gray1=cv2.imread("explore2.jpg",0)
img1=cv2.cvtColor(gray1,cv2.COLOR_BGR2RGB)
p.imshow(img1)
```
