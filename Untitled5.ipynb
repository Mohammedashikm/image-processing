```python
#Image processing libary
from skimage.feature import hog
from skimage import data, exposure
```

# READ IMAGE

```python
image=cv2.imread('nature3.jpg')
```

# HOG

```python
fd, hod_image = hog(image,orientations=9,pixels_per_cell=(8,8),cells_per_block=(2,2), visualize=True, channel_axis=2)
```

# RESCALING

```python
rescale=exposure.rescale_intensity(hod_image, in_range=(0,10))
```

# Plot

```python
figure,(a1,a2) =plt.subplots(1,2, figsize=(12,6), sharex=True, sharey=True)


a1.axis('off')
a1.imshow(image)
a1.set_title('Input image')

a2.axis('off')
a2.imshow(rescale)
a2.set_title('Histrogram of Oriented Gradients')
plt.show()
```

```python
import cv2
import imutils

# Initializing the HOG person
# detector
hog=cv2.HOGDescriptor()
hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())
# Reading the Image
image=cv2.imread('OIP.jpg')

#Resizing the Image
image =imutils.resize(image,width=min(400,image.shape[1]))

#Detectig all the in the 
#image that a pedestrians inside it
(regions,img) =hog.detectMultiScale(image,winStride=(2,2),padding=(4,4),scale=1.05)

# Drawing the regions in the Image
for(x, y, w, h) in regions:
    cv2.rectangle(image,(x,y),(x+w,y+h),(0,255,0),2)
    
# Showing the output Image
cv2.imshow("OIP.jpg",image)
cv2.waitKey(0)
            
cv2.destroyAllWindows()
            
```
